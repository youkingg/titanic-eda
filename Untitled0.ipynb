```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
df = pd.read_csv("train.csv")
df.head()  # 先頭5行を表示
# データの形（行数, 列数）
print(df.shape)

# 列名一覧
print(df.columns)

# 各列のデータ型と欠損値数
print(df.info())

# 基本統計量
print(df.describe())

# 欠損値の確認
print(df.isnull().sum())
# 生存者の割合
sns.countplot(x="Survived", data=df)
plt.show()

# 性別ごとの生存者割合
sns.countplot(x="Sex", hue="Survived", data=df)
plt.show()

# 年齢分布
sns.histplot(df["Age"].dropna(), bins=20, kde=True)
plt.show()
```
